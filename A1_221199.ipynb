Q1. Suppose you have a jar with 4 jelly beans: 2 are strawberry-flavored, 1 is blueberry-flavored, and 1 is cinnamon-flavored. You draw one jelly bean at random from the jar.

    a. What is the sample space for this experiment?

    b. We define event A as the jelly bean drawn is strawberry-flavored and event B as The jelly bean drawn is not cinnamon-flavored. What are the probabilities of events A and B?

    c. Are events A and B mutually exclusive? Why or why not?


```python
S = ['Strawberry', 'Strawberry', 'Blueberry', 'Cinnamon']
A = S.count('Strawberry') / len(S)
B = 1- S.count('Cinnamon') / len(S)
mutually_exclusive = False  # because there are strawberry-flavored jelly beans that are not cinnamon
print(f"Sample space: {S}")
print(f"Probability of event A (Strawberry): {A}")
print(f"Probability of event B (Not Cinnamon): {B}")
print(f"Are events A and B mutually exclusive? {mutually_exclusive}")
```

 Q2. Previously, we defined a **Python function P** to compute the probability of an event using the naive definition of probability. Generalize that function to compute the probability of events when they are
 not all equally likely. Use this new function to compute the probability of events A and B from the previous exercise.

 Hint: you can pass a third argument with the probability of each event.

```python
def P(event, sample_space, probabilities):
    """
    event: List of outcomes that define the event.
    sample_space: List of all possible outcomes.
    probabilities: List of probabilities corresponding to each outcome in the sample space.
    """
    # Create a dictionary to map each outcome to its probability
    outcome_prob = dict(zip(sample_space, probabilities))

    # Sum the probabilities of the outcomes that are in the event
    event_probability = sum(outcome_prob[outcome] for outcome in event)

    return event_probability

# Define the sample space and corresponding probabilities
sample_space = ['Strawberry', 'Strawberry', 'Blueberry', 'Cinnamon']
probabilities = [0.25, 0.25, 0.25, 0.25]  # Since there are 4 outcomes, each is equally likely (1/4)

# Define events A and B
event_A = ['Strawberry','Strawberry']
event_B = ['Strawberry', 'Strawberry', 'Blueberry']  # Not Cinnamon

# Calculate probabilities using the generalized function
prob_A = P(event_A, sample_space, probabilities)
prob_B = P(event_B, sample_space, probabilities)

# Output the results
print(f"Probability of event A (Strawberry): {prob_A}")
print(f"Probability of event B (Not Cinnamon): {prob_B}")
```

```python
# Define parameters for Gaussian distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Create Gaussian distribution object using scipy.stats
gaussian = norm(mu, sigma)

# Define range for plotting
x = np.linspace(gaussian.ppf(0.01), gaussian.ppf(0.99), 100)

# Plot PDF
plt.plot(x, gaussian.pdf(x), 'r-', lw=2, label='PDF')
plt.title("PDF of Gaussian Distribution")
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show()

# Plot CDF
plt.plot(x, gaussian.cdf(x), 'b-', lw=2, label='CDF')
plt.title("CDF of Gaussian Distribution")
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()

# Plot PPF
y = np.linspace(0.01, 0.99, 100)
plt.plot(y, gaussian.ppf(y), 'g-', lw=2, label='PPF')
plt.title("PPF of Gaussian Distribution")
plt.xlabel('Probability')
plt.ylabel('Quantile')
plt.legend()
plt.show()
```

```python
# Define parameters for BetaBinomial distribution
n = 10  # Number of trials
a = 2  # Alpha parameter
b = 2  # Beta parameter

# Create BetaBinomial distribution object using scipy.stats
beta_binomial = betabinom(n, a, b)

# Define range for plotting
x = np.arange(0, n + 1)

# Plot PMF
plt.vlines(x, 0, beta_binomial.pmf(x), colors='r', lw=5, label='PMF')
plt.title("PMF of BetaBinomial Distribution")
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()
plt.show()

# Plot CDF
plt.plot(x, beta_binomial.cdf(x), 'b-', lw=2, label='CDF')
plt.title("CDF of BetaBinomial Distribution")
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()

# Plot PPF
y = np.linspace(0.01, 0.99, 100)
plt.plot(y, beta_binomial.ppf(y), 'g-', lw=2, label='PPF')
plt.title("PPF of BetaBinomial Distribution")
plt.xlabel('Probability')
plt.ylabel('Quantile')
plt.legend()
plt.show()
```

Q4. Use PreliZ to compute the moments for the SkewNormal distribution for a different combination of parameters. Generate random samples of different sizes,
like 10, 100, and 1,000, and see if you can recover the values of the first two moments (mean and variance) from the samples. Report your obervations.

```python
import preliz as pz
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# Define the parameters for the SkewNormal distribution
loc = 0     # location parameter (mean)
scale = 1   # scale parameter (standard deviation)
shape = 5   # shape parameter (skewness)

# Create SkewNormal distribution using scipy.stats
skew_normal = skewnorm(a=shape, loc=loc, scale=scale)

# Theoretical moments
mean_theoretical = skew_normal.mean()
var_theoretical = skew_normal.var()

print(f"Theoretical mean: {mean_theoretical}")
print(f"Theoretical variance: {var_theoretical}")

# Function to calculate moments from samples
def calculate_moments(samples):
    mean_sample = np.mean(samples)
    var_sample = np.var(samples, ddof=1)  # Use ddof=1 for unbiased estimate
    return mean_sample, var_sample

# Generate random samples of different sizes
sample_sizes = [10, 100, 1000]
moments = {}

for size in sample_sizes:
    samples = skew_normal.rvs(size=size)
    mean_sample, var_sample = calculate_moments(samples)
    moments[size] = (mean_sample, var_sample)
    print(f"Sample size: {size}")
    print(f"Sample mean: {mean_sample}")
    print(f"Sample variance: {var_sample}")

# Plotting the PDF and sample histograms for visualization
x = np.linspace(skew_normal.ppf(0.01), skew_normal.ppf(0.99), 100)
plt.plot(x, skew_normal.pdf(x), 'r-', lw=2, label='PDF')
plt.title("PDF of SkewNormal Distribution")
plt.legend()
plt.show()

for size in sample_sizes:
    samples = skew_normal.rvs(size=size)
    plt.hist(samples, bins=30, density=True, alpha=0.6, label=f'Sample size {size}')
    plt.plot(x, skew_normal.pdf(x), 'r-', lw=2)
    plt.title(f"Histogram of Samples (size={size}) and PDF")
    plt.legend()
    plt.show()
```
